```python
import pandas as pd
df2 = pd.read_csv("/content/data_1.csv")
df2
```

```python
# from sqlalchemy.engine import result
import numpy as np

# Extract the lower and upper limit values for each nutrient level
lower_values = [list(df2.iloc[i, [1,3,5,7]]) for i in range(len(df2))]
upper_values = [list(df2.iloc[i, [2,4,6,8]]) for i in range(len(df2))]

# Create two lists for lower and upper values
lower_list = []
print(type(lower_list))
upper_list = []
column_name = ['N','P','K','pH']

for i in range(len(df2)):
    # Create sub-lists for lower and upper values
    lower_sublist = [df2.iloc[i,0], lower_values[i][0], lower_values[i][1], lower_values[i][2], lower_values[i][3]]
    upper_sublist = [df2.iloc[i,0], upper_values[i][0], upper_values[i][1], upper_values[i][2], upper_values[i][3]]
    lower_list.append(lower_sublist)
    upper_list.append(upper_sublist)
print("lower_limit", lower_list)
print("upper_limit", upper_list)

# test data
test_case = (3.0, 0.20,1.50,6.2)
print("check")
print(type(test_case))
print("test_case = ",test_case)
import csv
# test_case = []
# with open("/content/test_1.csv") as file_obj:
#     reader_obj = csv.reader(file_obj)
#     for row in reader_obj:
#         test_case.append(row)



results = []
for i in range(len(test_case)):
    temp = []

    for j in range(len(lower_list)):

        if np.all(lower_list[j][i+1] <= test_case[i]) and np.all(test_case[i] <= upper_list[j][i+1]):
            temp.append(lower_list[j][0])

# print(temp)
    results.append(temp)
print(results)
print(type(results))
print(results[2])
```

```python
import array
fruit_count = {}

# count the number of occurrences of each fruit
for sublist in results:
    for fruit in sublist:
        # print(fruit)
        if fruit in fruit_count:
            fruit_count[fruit] += 1
        else:
            fruit_count[fruit] = 1

# create lists of fruits occurring 4 and 3 times
print(fruit_count)
four_times = []
three_times = []
two_times = []
# one_time = []
for fruit, count in fruit_count.items():
    if count == 4:
        four_times.append(fruit)
    elif count == 3:
        three_times.append(fruit)
    elif count == 2:
        two_times.append(fruit)
    # else:
    #     one_time.append(fruit)

print("Fruits occurring 4 times:", four_times)
print("Fruits occurring 3 times:", three_times)
print("Fruits occurring 2 times:", two_times)
# print("Fruits occurring 1 time:", one_time)

```
